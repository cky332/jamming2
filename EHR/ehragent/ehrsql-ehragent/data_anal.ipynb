```python
import json
import jsonlines
```

```python
# read file line by line
dataset = "mimic_iii"
file_path = "/<YOUR_DATA_PATH>/{}/valid_filtered_20240104.jsonl".format(dataset)
with jsonlines.open(file_path) as reader:
    lines = list(reader)
print("Total number of valid queries: {}".format(len(lines)))
none_answers = 0
new_lines = []
for line in lines:
    if "None" in line["answer"]:
        none_answers += 1
    else:
        new_lines.append(line)
print("Number of queries with None answers: {}".format(none_answers))
```

```python
output_path = "/<YOUR_DATA_PATH>/{}/valid_20240104.json".format(dataset)
with open(output_path, "w") as writer:
    json.dump(new_lines, writer, indent=4)

output_path = "/<YOUR_DATA_PATH>/{}/valid_20240104.jsonl".format(dataset)
with jsonlines.open(output_path, "w") as writer:
    writer.write_all(new_lines)
```
